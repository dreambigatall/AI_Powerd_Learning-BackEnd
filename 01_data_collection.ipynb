```python
import requests
import pandas as pd
import time
print("Libraries imported successfully.")
```

```python
# GitHub API Configuration
technologies = [
    "facebook/react",
    "vercel/next.js",
    "nestjs/nest",
    "golang/go",
    "rust-lang/rust",
    "expressjs/express",
    "django/django",
    "fastapi/fastapi",
    "spring-projects/spring-boot",
    "laravel/laravel"
]

def fetch_repo_data(repo_path):
    url = f"https://api.github.com/repos/{repo_path}"
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Failed to fetch {repo_path}: {response.status_code}")
        return None
```

```python
# Collect Data
results = []

for repo in technologies:
    print(f"Fetching data for {repo}...")
    data = fetch_repo_data(repo)
    if data:
        results.append({
            "technology": data["name"],
            "stars": data["stargazers_count"],
            "forks": data["forks_count"],
            "watchers": data["subscribers_count"],
            "issues": data["open_issues_count"]
        })
    # Small sleep to be respectful to API rate limits
    time.sleep(0.5)

df = pd.DataFrame(results)
display(df.head(10))
```

```python
# Save CSV
df.to_csv("tech_data.csv", index=False)
print("Data saved to tech_data.csv")

# Quick Analysis
print("\nTop Repo by Stars:")
display(df.sort_values("stars", ascending=False).head(1))
```
